```python
import pandas
```

```python
#model parmeters

Revenue_Growth_Factor = 1.18 # % of prior period Revenue
Ebitda_margin = 0.45 # % of current Revenue
Tax_pct= 0.21 # % of current EDBITDA
Cap_Exp = 0.03 # of current Revenue
Work_Cap_Increase = 0.06 # % of Revenue
Discount_Factors = 0.94 #(6%)
Ebitda_Exit_Multiple = 32.70
```

```python
dates = pandas.date_range(start = '2022', end = '2032',freq = 'y')
dates
```

```python
model_df = pandas.DataFrame(index = dates)
model_df
```

```python
# Revenue :

start_revenue = 222 

revenue_projection = [start_revenue * Revenue_Growth_Factor]

for i in range(1,len(dates)):
    revenue_projection.append(revenue_projection[i-1]* Revenue_Growth_Factor)   

model_df['Revenue']= revenue_projection
model_df
```

```python
# EBITDA
model_df['Ebitda'] = model_df['Revenue'] * Ebitda_margin
model_df
```

```python
# Taxes
model_df['Taxes'] = model_df['Ebitda'] *  -Tax_pct
model_df
```

```python
# Capital Expendutiers
model_df['Cap_Exp'] = model_df['Revenue'] *  - Cap_Exp
model_df
```

```python
# Work_Cap_Increase

model_df['Work_Cap'] = model_df['Revenue'] * -Work_Cap_Increase
model_df
```

```python
model_df['Cash_Flow'] = model_df.drop(columns =['Revenue']).sum(axis = 1)
model_df
```

```python
model_df['Discount_Factors'] = [Discount_Factors**i for i in range(1,len(dates)+1)]
model_df
```

```python
model_df['PV_of_FCF']= model_df['Discount_Factors'] * model_df['Cash_Flow'].values
model_df.round(2)
```

```python
PV_of_projected_FCF = model_df['PV_of_FCF'].sum()
PV_of_projected_FCF.round(0)
```

```python
model_df.iloc[-1]['Revenue']*Revenue_Growth_Factor
```

```python
Ebitda_2032 = round(model_df.iloc[-1]['Revenue']*1.18*0.45)

print(f"""
Our projected Present Value of Projected Future Cash Flows are: {PV_of_projected_FCF.round(0)}.
We are providing revenue guidance in 2032 to be: {round(model_df.iloc[-1]['Revenue']*Revenue_Growth_Factor)}.
With an estimated EBITDA around: {round(model_df.iloc[-1]['Revenue']*1.18*0.45)}

The terminal value a this time given a multiple of {Ebitda_Exit_Multiple} and the above EBITDA for 2032, would equal: {round(Ebitda_Exit_Multiple*Ebitda_2032)})
""")
```
